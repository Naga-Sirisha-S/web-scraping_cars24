```python
# Additional statistics
mean_price = df['Price (in lakhs)'].mean()
median_price = df['Price (in lakhs)'].median()
mode_price = df['Price (in lakhs)'].mode()[0]
std_price = df['Price (in lakhs)'].std()

mean_kilometers = df['Kilometers_Driven'].mean()
median_kilometers = df['Kilometers_Driven'].median()
mode_kilometers = df['Kilometers_Driven'].mode()[0]
std_kilometers = df['Kilometers_Driven'].std()

mean_year = df['Year_of_Manufacture'].mean()
median_year = df['Year_of_Manufacture'].median()
mode_year = df['Year_of_Manufacture'].mode()[0]
std_year = df['Year_of_Manufacture'].std()

print(f"Price (in lakhs) - Mean: {mean_price}, Median: {median_price}, Mode: {mode_price}, Std Dev: {std_price}")
print(f"Kilometers Driven - Mean: {mean_kilometers}, Median: {median_kilometers}, Mode: {mode_kilometers}, Std Dev: {std_kilometers}")
print(f"Year of Manufacture - Mean: {mean_year}, Median: {median_year}, Mode: {mode_year}, Std Dev: {std_year}")
```

#### Observation_3 : The average price of the cars in the dataset is approximately 7.09 lakhs with stan.dev of 3 lakhs.The car prices vary a lot, but many of them are on the cheaper side. Some expensive cars raise the average price, showing that the dataset includes both budget-friendly and luxury models.

```python
# Frequency distribution for categorical columns
company_name_distribution = df['Company_Name'].value_counts()
fuel_type_distribution = df['Fuel_Type'].value_counts()
transmission_distribution = df['Transmission'].value_counts()
Location_distribution = df['Location'].value_counts()

print("Frequency Distribution for Company Name:\n", company_name_distribution)
print("---------------------------------->")
print("Frequency Distribution for Fuel Type:\n", fuel_type_distribution)
print("---------------------------------->")
print("Frequency Distribution for Transmission:\n", transmission_distribution)
print("---------------------------------->")
print("Frequency Distribution for Location:\n", Location_distribution)

```


#### Observation_4: Frequency distribution for the company_name shows that the Hyundai is the most represented brand in the dataset, with 1,255 cars. This indicates Hyundai's strong presence and popularity in the market, whereas jeep with 53 cars, indicates a smaller market share.
#### Petrol vehicles are more preferred choice whereas electric vehicles are barely present.
#### Manual transmissions are more prevalent in the dataset, suggesting they are the preferred choice, while automatic transmissions are also significant, reflecting a growing trend towards convenience and ease of driving.
#### The dataset shows that Bangalore has the highest number of cars, indicating a strong market presence.

```python
# Calculating means
mean_price = df['Price (in lakhs)'].mean()
mean_kilometers = df['Kilometers_Driven'].mean()
mean_year = df['Year_of_Manufacture'].mean()

# Histograms with KDE and mean lines for 'Price (in lakhs)', 'Kilometers_Driven', and 'Year_of_Manufacture'
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
sns.histplot(df['Price (in lakhs)'], bins=20, kde=True, color='blue')
plt.axvline(mean_price, color='black', linestyle='--', linewidth=1)
plt.title('Price Distribution')
plt.xlabel('Price (in lakhs)')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

plt.subplot(1, 3, 2)
sns.histplot(df['Kilometers_Driven'], bins=20, kde=True, color='green')
plt.axvline(mean_kilometers, color='black', linestyle='--', linewidth=1)
plt.title('Kilometers Driven Distribution')
plt.xlabel('Kilometers Driven')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
sns.histplot(df['Year_of_Manufacture'], bins=20, kde=True, color='Red')
plt.axvline(mean_year, color='black', linestyle='--', linewidth=1)
plt.title('Year of Manufacture Distribution')
plt.xlabel('Year of Manufacture')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### Obsevation_5:In 1 subplot the black dashed line represents the mean price, which is slightly to the right of the peak. This suggests that while most cars are priced below the mean, a few expensive cars raise the average price.
### In 2 subplot the distribution of kilometers driven is mostly balanced, with most cars having moderate mileage and a few having high mileage.
### In 3 subplot the histogram for the year of manufacture is mostly to the left, showing that most cars are relatively new.

```python
# Boxplots to identify outliers in 'Price (in lakhs)', 'Kilometers_Driven', and 'Year_of_Manufacture'
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
sns.boxplot(y=df['Price (in lakhs)'], color='orange')
plt.title('Boxplot of Price (in lakhs)')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['Kilometers_Driven'], color='purple')
plt.title('Boxplot of Kilometers Driven')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['Year_of_Manufacture'], color='green')
plt.title('Boxplot of Year of Manufacture')

plt.tight_layout()
plt.show()
```

#### Observation_6: In 1 boxplot,it shows several outliers on the higher end, indicating that there are a few cars with significantly higher prices compared to the majority.
#### In 2 boxplot,it shows a few cars with unusually high mileage compared to the rest.

```python
# Bar charts to visualize the frequency of each 'Company_Name', 'Fuel_Type', and 'Transmission'
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
df['Company_Name'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Cars by Company')
plt.xlabel('Company Name')
plt.ylabel('Count')

plt.subplot(1, 3, 2)
df['Fuel_Type'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')

plt.subplot(1, 3, 3)
df['Transmission'].value_counts().plot(kind='bar', color='salmon')
plt.title('Number of Cars by Transmission')
plt.xlabel('Transmission')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

#### Observation_7:  1.subplot shows Hyundai has the most cars, indicating it's the most popular brand, while other brands have fewer cars.
#### 2. subplot shows that most cars run on petrol, with fewer diesel and CNG cars, and electric cars are very uncommon.
#### 3. subplot shows more cars have manual transmissions than automatic ones, showing people prefer manual cars.

```python
fuel_color_map = cm.get_cmap('Set1')
transmission_color_map = cm.get_cmap('Paired')

# Stacked bar chart for Fuel_Type distribution
company_fuel = pd.crosstab(df['Company_Name'], df['Fuel_Type'])
company_fuel.plot(kind='bar', stacked=True, figsize=(14, 6), colormap=fuel_color_map)
plt.title('Distribution of Fuel Type across Companies')
plt.xlabel('Company Name')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()

# Stacked bar chart for Transmission distribution
company_transmission = pd.crosstab(df['Company_Name'], df['Transmission'])
company_transmission.plot(kind='bar', stacked=True, figsize=(14, 6), colormap=transmission_color_map)
plt.title('Distribution of Transmission Type across Companies')
plt.xlabel('Company Name')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

#### Observation_8 : In 1. barchart shows what fuels each company uses, with Hyundai and other big brands offering petrol, diesel, and CNG.
#### In 2 barchart shows how many cars have manual versus automatic transmissions for each company, with Hyundai and other brands mostly offering manual transmissions.

```python
# Scatter plots to visualize the relationship between 'Price (in lakhs)' and 'Kilometers_Driven'
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(df['Kilometers_Driven'], df['Price (in lakhs)'], color='skyblue', edgecolor='black')
plt.title('Price vs Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price (in lakhs)')

# Scatter plots to analyze 'Price (in lakhs)' against 'Year_of_Manufacture'
plt.subplot(1, 2, 2)
plt.scatter(df['Year_of_Manufacture'], df['Price (in lakhs)'], color='lightgreen', edgecolor='black')
plt.title('Price vs Year of Manufacture')
plt.xlabel('Year of Manufacture')
plt.ylabel('Price (in lakhs)')

plt.tight_layout()
plt.show()
```

#### Observation_9 : 1. plot shows cars with more kilometers driven have lower prices.
#### 2. plot shows that newer cars usually cost more, with prices going up for more recent models.

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Correlation matrix for numeric columns
corr_matrix = df[numeric_cols].corr()

# Plotting the heatmap
plt.figure(figsize=(5, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Numeric Variables')
plt.show()
```

### Observation_10: The heatmap displays how numeric variables are related to each other. Strong correlations are shown by values close to 1 or -1, while values near 0 indicate weak or no relationship.

```python
# Defining custom colors excluding pink
fuel_type_colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff6666']
transmission_colors = ['#c2c2f0', '#80bfff']

# Pie charts for a quick view of the proportion of different 'Fuel_Type' and 'Transmission'
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

df['Fuel_Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=fuel_type_colors, ax=axes[0])
axes[0].set_title('Proportion of Fuel Types')
axes[0].set_ylabel('')

df['Transmission'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=transmission_colors, ax=axes[1])
axes[1].set_title('Proportion of Transmission Types')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
custom_palette = {
    'Petrol': '#1f77b4',   # Blue
    'Diesel': '#ff7f0e',   # Orange
    'CNG': '#2ca02c',      # Green
    'Electric': '#d62728'  # Red
}
plt.figure(figsize=(15, 6))

# Regression plot with separate lines for each Fuel_Type
sns.lmplot(data=df, x='Kilometers_Driven', y='Price (in lakhs)', hue='Fuel_Type', palette=custom_palette,
           height=6, aspect=1.5, scatter_kws={'s':50}, line_kws={'linewidth':2})

plt.title('Price vs Kilometers Driven by Fuel Type')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price (in lakhs)')
plt.grid(True)

plt.show()
```

```python
# Calculating the average price by year
avg_price_by_year = df.groupby(['Year_of_Manufacture', 'Company_Name'])['Price (in lakhs)'].mean().reset_index()
avg_price_by_year.rename(columns={'Price (in lakhs)': 'Avg_Price (in lakhs)'}, inplace=True)
print(avg_price_by_year)


# Creating a line plot for average price by year
sns.lineplot(data=avg_price_by_year, x='Year_of_Manufacture', y='Avg_Price (in lakhs)', hue='Company_Name', marker='o')

plt.title('Average Price of Vehicles by Year')
plt.xlabel('Year of Manufacture')
plt.ylabel('Average Price (in lakhs)')
plt.legend(title='Company Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

plt.show()
```

### Observation_10: The plot shows how the average price of cars changes each year for different companies. Some companies have higher prices for newer models, while others have more consistent prices.
